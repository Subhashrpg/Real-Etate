```python
# price column ============================================================================================
df[df["price"] == "Price on Request"]
```

```python
df = df[df["price"] != "Price on Request"]
```

```python
df.head()
```

```python
def fix_price(row):
    if type(row) == float:
        return row
    else:
        if row[1] == 'Lac':
            return round(float(row[0])/100,2)
        else:
            return round(float(row[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(fix_price)
```

```python
df.head()
```

```python
# price_per_sqft column ====================================================================================
df["price_per_sqft"].value_counts().sample(20)
```

```python
df["price_per_sqft"] = df["price_per_sqft"].str.split("/").str.get(0).str.strip().str.replace("₹","").str.strip().str.replace(",","").astype("float")
```
