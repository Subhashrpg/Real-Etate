```python
# imports
import pandas as pd
import numpy as np
import re
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv("datasets/Gurgaon/Independent House/independent-house - independent-house.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# duplicate values
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(columns=["property_id","link"],inplace=True)
```

```python
df.head()
```

```python
# rename columns
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
df.head()
```

```python
# society column
df.society.value_counts().shape, df.society.value_counts()
```

```python
df.society = df.society.apply(lambda name: re.sub(r"\d+(\.\d+)?\s?★", "", str(name))).str.strip().str.lower()
```

```python
df.society.value_counts().shape
```

```python
df.society = df.society.str.replace("nan","independent")
```

```python
df.head()
```

```python
# price column
temp_df = df[df.price != "Price on Request"]
```

```python
def fix_price(row):
    if type(row) == float:
        return row
    else:
        if row[1] == 'Lac':
            return round(float(row[0])/100,2)
        else:
            return round(float(row[0]),2)
```

```python
temp_df.price = temp_df.price.str.split(" ").apply(fix_price)
```

```python
temp_df.head(2)
```

```python
# price_per_sqft column
```

```python
temp_df.price_per_sqft.value_counts()
```

```python
temp_df.price_per_sqft = temp_df.price_per_sqft.str.split("/").str.get(0).str.replace("₹","").str.replace(",","").str.strip().astype("float")
```

```python
temp_df.sample(2)
```

```python
# bedroom column
temp_df.bedRoom.value_counts()
```

```python
temp_df.bedRoom.isnull().sum()
```

```python
temp_df = temp_df[~temp_df.bedRoom.isnull()]
temp_df.head()
```

```python
temp_df.bedRoom = temp_df.bedRoom.str.split(" ").str.get(0).str.strip().astype("int")
```

```python
temp_df.head()
```

```python
# bathroom column
temp_df.bathroom.isnull().sum(),temp_df.bathroom.value_counts()
```

```python
temp_df.bathroom = temp_df.bathroom.str.split(" ").str.get(0).str.strip().astype("int")
```

```python
temp_df.sample()
```

```python
# balcony column
temp_df.balcony.value_counts()
```

```python
temp_df.balcony.isnull().sum()
```

```python
temp_df.balcony = temp_df.balcony.str.split(" ").str.get(0).str.strip().replace("No","0")
```

```python
temp_df.sample()
```

```python
# additinalroom column
temp_df.additionalRoom.value_counts()
```

```python
temp_df.additionalRoom.isnull().sum()
```

```python
temp_df.additionalRoom.fillna("not available", inplace=True)
```

```python
temp_df.additionalRoom = temp_df.additionalRoom.str.lower()
```

```python
temp_df.sample()
```

```python
# noOffloors column
temp_df.noOfFloor.isnull().sum(),temp_df.noOfFloor.value_counts()
```

```python
temp_df.noOfFloor= temp_df.noOfFloor.str.split(" ").str.get(0)
```

```python
temp_df.rename(columns= {"noOfFloor": "floor_num"}, inplace=True)
```

```python
temp_df.head(3)
```

```python
# facing column
temp_df['facing'].fillna('NA',inplace=True)
```

```python
temp_df['area'] = round((temp_df['price']*10000000)/temp_df['price_per_sqft'])
```

```python
temp_df.insert(loc=1,column='property_type',value='house')
```
